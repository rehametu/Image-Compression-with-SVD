* In the project 64x64 images are used. To achieve better performance, PCA is used for image compersion. Before use of PCA onn the project, number of principal components are determined by trial and error. 'lion1.jpg' is used for this example also this image used for test on the project. Principal component number is selected as 45 (looks fine :) ).

* Resource [link](https://www.askpython.com/python/examples/principal-component-analysis-for-image-data) is added.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
img = cv2.imread('lion1.jpg') #you can use any image you want.
```

```python
img = cv2.imread('lion1.jpg')
RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(RGB)
plt.title('Original Image',
          fontsize=30,
          color='orange')
plt.show()
```

```python
blue,green,red = cv2.split(img)
```

```python
blue.shape
```

```python
pca = PCA(45)
```

```python
red_transformed = pca.fit_transform(red)
red_inverted = pca.inverse_transform(red_transformed)

green_transformed = pca.fit_transform(green)
green_inverted = pca.inverse_transform(green_transformed)

blue_transformed = pca.fit_transform(blue)
blue_inverted = pca.inverse_transform(blue_transformed)
```

```python
img_compressed = (np.dstack((red_inverted, green_inverted, blue_inverted))).astype(np.uint8)
```

```python
plt.title('Test Picture with 2 Principle Component\n',
          fontsize=30,
          color="red")
plt.imshow(img_compressed) #2
```

```python
plt.title('Test Picture with 10 Principle Component\n',
          fontsize=30,
          color="red")
plt.imshow(img_compressed) #10
```

```python
plt.title('Test Picture with 50 Principle Component\n',
          fontsize=30,
          color="red")
plt.imshow(img_compressed) #50
```

```python
plt.title('Test Picture with 45 Principle Component\n',
          fontsize=30,
          color="green")
plt.imshow(img_compressed) #45
```
